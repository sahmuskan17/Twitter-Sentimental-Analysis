```python
# Normalize text data by converting to lowercase and removing punctuation
def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(f"[{re.escape(string.punctuation)}]", "", text)  # Remove punctuation
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    return text

df['text'] = df['text'].apply(preprocess_text)
df['selected_text'] = df['selected_text'].apply(preprocess_text)
```

### Text Preprocessing

Text preprocessing is a critical step in natural language processing (NLP) that transforms raw text into a cleaner, standardized format for analysis. It involves converting text to lowercase, removing punctuation, and eliminating stop words. Additional steps include tokenization, stemming, lemmatization, and handling special characters or numbers. These processes help in reducing noise and extracting meaningful information, making the text data suitable for machine learning models and other analytical tasks

```python
import nltk 
nltk.download('stopwords') 
nltk.download('punkt')
nltk.download('wordnet')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
# Initialize stopwords and lemmatizer
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
# Define text preprocessing function
def preprocess_text(text):
    # Tokenize text
    tokens = word_tokenize(text)
    # Remove stopwords and non-alphabetic tokens, and apply lemmatization
    tokens = [lemmatizer.lemmatize(word.lower()) for word in tokens if word.isalpha() and word.lower() not in stop_words]
    # Join tokens back to a single string
    return ' '.join(tokens)
```

```python
# Apply the preprocessing function to the review text
df['processed_text'] = df['selected_text'].apply(preprocess_text)
```

```python
# Display the first few rows of the dataset to check the processed text
df[['selected_text', 'processed_text']].head()
```

### Converting text into Vectors
TF-IDF calculates that how relevant a word in a series or corpus is to a text. The meaning increases proportionally to the number of times in the text a word appears but is compensated by the word frequency in the corpus (data-set). We will be implementing this with the code below.

```python
# Initialize the TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=5000) 
```

```python
# Fit and transform the processed text
tfidf_matrix = tfidf_vectorizer.fit_transform(df['processed_text'])
```

```python
# Convert the TF-IDF matrix to a DataFrame
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
```

```python
# Add sentiment labels to the TF-IDF DataFrame
tfidf_df['sentiment'] = df['sentiment']
```

```python
# Display the first few rows of the TF-IDF DataFrame
tfidf_df.head()
```

### Model training, Evaluation, and Prediction
 We can now explore any machine learning model to train the data

```python
# Get X (tweets) and y (labels)
labenc = LabelEncoder()
labels = labenc.fit_transform(df['sentiment'])
tweets = df['processed_text']

# Train test split
X_train, X_test, y_train, y_test = train_test_split(tweets, labels, test_size=0.3, random_state=42, stratify=labels)

# Vectorize the tweets
vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
# Fit Support Vector Classifier (slow)
model = SVC()
model.fit(X_train_vec, y_train)

# Predict and evaluate
y_pred = model.predict(X_test_vec)
print("Train accuracy:", model.score(X_train_vec,y_train))
print("Test accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred, target_names=sentiments))
```

```python
top_words = pd.Series(np.concatenate([x.split() for x in df['processed_text']])).value_counts()
top_words.head(20)
```

```python
top_words_top = top_words[:1000]
top_labeled = {}

# SLOW - Count how many positive, negative, neutral tweets each top word appears in
for word, freq in top_words_top.items():
  top_labeled[word] = [0,0,0] # pos neg neut
  for tweet, label in zip(df['processed_text'],df['sentiment']):
    if word in tweet:
      if label == 'positive':
        top_labeled[word][0] += 1
      elif label == 'negative':
        top_labeled[word][1] += 1
      else:
        top_labeled[word][2] += 1

# Label each word the class corresponding to the highest number in top_labeled[word]
for word, counts in top_labeled.items():
  if counts[0] >= counts[1] and counts[0] >= counts[2]:
    top_labeled[word] = [counts, 'positive']
  elif counts[1] >= counts[0] and counts[1] >= counts[2]:
    top_labeled[word] = [counts, 'negative']
  else:
    top_labeled[word] = [counts, 'neutral']
```

```python
# Generate dict like top_words_top but one for each class
top_pos_freq = {}
top_neg_freq = {}
top_neut_freq = {}

for key, val in top_labeled.items():
  if val[1] == 'positive':
    top_pos_freq[key] = top_words_top[key]
  elif val[1] == 'negative':
    top_neg_freq[key] = top_words_top[key]
  else:
    top_neut_freq[key] = top_words_top[key]
```

```python
def plot_word_cloud_dict_colormapped(d, color_fun, title):
  plt.style.use("dark_background")
  wordcloud = WordCloud(collocations=False,width=1600, height=800, prefer_horizontal=1).generate_from_frequencies(d)
  wordcloud.recolor(color_func=color_fun)
  # Plot the wordcloud image
  plt.figure(figsize=(14,10))
  plt.imshow(wordcloud)
  plt.axis("off")
  plt.title(title)
  plt.show()
```

```python
# Generate the positive wordcloud in green
color_func_1 = lambda **kwargs: "#04d104"

plot_word_cloud_dict_colormapped(top_pos_freq, color_func_1, 'Most frequent words in tweets labeled positive')
```

```python
# Generate the positive wordcloud in nutral
color_func_1 = lambda **kwargs: "#d7bda3"

plot_word_cloud_dict_colormapped(top_neut_freq, color_func_1, 'Most frequent words in tweets labeled nutral')
```

```python
# Generate the negative wordcloud in red
color_func_2 = lambda **kwargs: "#d20303"

plot_word_cloud_dict_colormapped(top_neg_freq, color_func_2, 'Most frequent words in tweets labeled negative')
```
